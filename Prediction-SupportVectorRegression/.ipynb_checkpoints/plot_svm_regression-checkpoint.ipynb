# Destek Vektör İle Tahmin
SVR, normal doğrusal regresyondan farklı olarak verileri sadece doğrusal değil, çizgisel olmayan şekillerde de öğrenebilir.

linear: Doğru çizerek çözüm.

poly: Parabollerle çözüm.

rbf: En esnek yaklaşım, karmaşık şekillerde çözüm.

```python
"""
===================================================================
Support Vector Regression (SVR) using linear and non-linear kernels
===================================================================

Toy example of 1D regression using linear, polynomial and RBF kernels.

"""
print(__doc__)

import numpy as np
from sklearn.svm import SVR
import matplotlib.pyplot as plt

###############################################################################
# Generate sample data - Örnek veri üretimi (Bağımsız Değişken)
X = np.sort(5 * np.random.rand(40, 1), axis=0) # 0-5 arasında 40 tane rastgele sayı üretip küçükten büyüğe sırala
y = np.sin(X).ravel() # hedef değişken üretildi (dalgalı eğri elde etmek için)

###############################################################################
# Add noise to targets - Veriye gürültü ekleme
y[::5] += 3 * (0.5 - np.random.rand(8))
# Her 5. elemanın değerine rastgele sapma (hata) ekleniyor.
# Amaç: Gürültü (noisy) veri simülasyonu yapmak
# The random module's rand() method 

###############################################################################
# Fit regression model
# RBF kernel → Eğri çizen model
svr_rbf = SVR(kernel='rbf', C=1e3, gamma=0.1)
# Linear kernel → Düz çizgi
svr_lin = SVR(kernel='linear', C=1e3)
# Polynomial kernel → Eğri (derece 2 polinom)
svr_poly = SVR(kernel='poly', C=1e3, degree=2)

# Modelleri eğit ve tahmin yap
y_rbf = svr_rbf.fit(X, y).predict(X)
y_lin = svr_lin.fit(X, y).predict(X)
y_poly = svr_poly.fit(X, y).predict(X)

###############################################################################
# look at the results - Grafik ile Sonuçları Gösterme
plt.scatter(X, y, c='k', label='data') # Verinin gerçek hali (scatter - siyah noktalar)
# plt.hold('on') # bu satır matplotlib'in eski sürümlerine ait, artık kullanılmıyor.
# Modellerin tahmin ettiği çizgileri çiziyoruz
plt.plot(X, y_rbf, c='g', label='RBF model')
plt.plot(X, y_lin, c='r', label='Linear model')
plt.plot(X, y_poly, c='b', label='Polynomial model')

plt.xlabel('data')
plt.ylabel('target')
plt.title('Support Vector Regression')
plt.legend() # Etiketleri göster
plt.show() # Grafiği göster
```
