### Support Vector Regression (SVR)
- SVR'nin Özelliği:
  SVR, tahmin yaparken verileri "bir doğruya en iyi şekilde uydurmak" yerine şöyle der:
  "Verilerin %90'ı bu tolerans aralığında kalmalı ama kenardaki uç değerler (outlier) modelimi çok da etkilememeli."
   Yani SVR, aşırı esnek veya sert olmayan dengeleyici bir tahmin sunar.
- Support Vector Regression, scaler ile kullanılması zorunludur.
- Marjinal (aykırı) verilere karşı dayanıklılığı olmadığı için scaler kullanılmalı!

### SVR'nin Avantajı Ne?
- Outlier’lara karşı daha dayanıklıdır.
- Verinin doğrusallığına çok bağımlı değildir (özellikle rbf kernel ile).
- Küçük veri setlerinde bile güzel sonuçlar verir.

### Nasıl Çalışır?
- SVR, doğrusal bir çizgi yerine, verilerle bir tüp (margin/tolerans aralığı) oluşturur.
- Bu tüpün dışındaki veriler "destek vektörü" (support vector) sayılır.
- Model bu destek vektörlere göre eğitilir.
- Özellikle karmaşık veri yapılarında kernel='rbf' gibi çekirdek fonksiyonları kullanılarak doğrusal olmayan tahminler yapılabilir.

### Ölçekleme Neden Gerekli? 
- X (tecrübe): 1-10 arasında
- Y (maaş): 1500 - 15000 arasında
  
Bu durumda model sadece maaşlara odaklanır. Bu yüzden StandardScaler() ile ortalama 0, standart sapma 1 olacak şekilde veriler normalize edilir.

|Model |Esneklik |Outliner Dayanıklılığı |Karmaşık Veri |
|:-----|:----:|----:|----:|
|Linear Regression |Düşük |Düşük |Zayıf  |
|Polynomial Regression |Orta-Yüksek |Düşük |İyi ama overfit |
|SVR(RBF) |Yüksek |Yüksek |Çok iyi|

### Polynomial vs SVR
- Polynomial Regression: Özellikle eğri gibi dağılımlarda esnek olur ama overfit riski var.
- SVR (özellikle RBF kernel): Genelde daha yumuşak bir eğri çizer. Overfit riski daha az, genellemeye daha uygun.

# Linear Regression: 
- __Düz Çizgi -> Verinin genel eğilimini takip eder.__
# Polynomial Regression 
- __Eğri Çizgi -> Veriye daha iyi uyum sağlar ama bazen fazla...__
# SVR
- __Margin Aralığı -> Hata toleransı tanır, uç değer varsa umursamayabilir.__

```python
# MODEL KURULUM - GRAFİK - TAHMİN
# 1. Kütüphaneler
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd

# 2. Veri Yükleme
veriler = pd.read_csv('maaslar.csv')

# X = Bağımsız Değişken (Tecrübe), Y = Bağımlı Değişken (Maaş)
x = veriler.iloc[:,1:2] # Tecrübe sütunu
y = veriler.iloc[:,2:] # Maaş sütunu
X = x.values
Y = y.values

print(x)
print(y)

# 3. Linear Regression
from sklearn.linear_model import LinearRegression
lin_reg = LinearRegression()
lin_reg.fit(X,Y) # Modeli veriye fit et (eğit)

plt.scatter(X,Y,color='red') # Gerçek veriyi çiz
plt.plot(x,lin_reg.predict(X), color = 'blue') # Modelin tahmin ettiği çizgiyi çiz
plt.title("Linear Regression: Tecrübeye Göre Maaş Tahmini")
plt.xlabel("Tecrübe (Yıl)")
plt.ylabel("Maaş")
plt.show()

# 4. Polynomial Regression (degree=2) 
from sklearn.preprocessing import PolynomialFeatures
poly_reg = PolynomialFeatures(degree = 2) # 2. Dereceden Polinom Eğri
x_poly = poly_reg.fit_transform(X) # X verisini polinom hale çevir.
print(x_poly)
lin_reg2 = LinearRegression()
lin_reg2.fit(x_poly,y) # Polinom modele fit et.
plt.scatter(X,Y,color = 'red')
plt.plot(X,lin_reg2.predict(poly_reg.fit_transform(X)), color = 'blue')
plt.title("Polynomial Regression (Derece=2): Tecrübeye Göre Maaş Tahmini")
plt.xlabel("Tecrübe (Yıl)")
plt.ylabel("Maaş")
plt.show()

""" 
    Polinom regresyonda derece (degree), modelin veriye ne kadar kıvrımlı bir eğri çizeceğini belirler.
    Dikkat: Derece artarsa = Overfitting riski!
    Derece düşükse: Model yeterince öğrenemez -> Underfitting
    Derece yüksekse: Model ezber yapar -> Overfitting
    Yani her noktanon üstünden geçeyim der -> Ama yeni veri geldiğinde başarısız olur.
"""
print(__doc__)
# 5. Polynomial Regression (degree = 4)
from sklearn.preprocessing import PolynomialFeatures # 4. Dereceden polinom
poly_reg = PolynomialFeatures(degree = 4)
x_poly = poly_reg.fit_transform(X)
print(x_poly)
lin_reg2 = LinearRegression()
lin_reg2.fit(x_poly,y)
plt.scatter(X,Y,color = 'red')
plt.plot(X,lin_reg2.predict(poly_reg.fit_transform(X)), color = 'blue')
plt.title("Polynomial Regression (Derece=4): Tecrübeye Göre Maaş Tahmini")
plt.xlabel("Tecrübe (Yıl)")
plt.ylabel("Maaş")
plt.show()
#TODO: Grid Search (Otomatik Seçim): Makine otomatik dener en iyi degree'yi sistem bulur.

# Tahminler
# 6. Linear Regression ile Tahmin
print("Linear Regression ile 11 yıl tecrübeye sahip birinin tahmini maaşı: {:.2f} TL".format(lin_reg.predict([[11]])[0][0]))
print("Linear Regression ile 6.6 yıl tecrübeye sahip birinin tahmini maaşı: {:.2f} TL".format(lin_reg.predict([[6.6]])[0][0]))

# 7. Polynomial Regression ile Tahmin
print("Polynomial Regression (derece=4) ile 11 yıl tecrübeye sahip birinin tahmini maaşı: {:.2f} TL".format(lin_reg2.predict(poly_reg.fit_transform([[11]]))[0][0]))
print("Polynomial Regression (derece=4) ile 6.6 yıl tecrübeye sahip birinin tahmini maaşı: {:.2f} TL".format(lin_reg2.predict(poly_reg.fit_transform([[6.6]]))[0][0]))

# 8. SVR - Support Vector Regression
# Not: SVR mesafe bazlı çalıştığı için verilerin ölçeklenmesi gerekir. Veriler aynı ölçekte olmazsa anlamlı tahminler yapamaz.
from sklearn.preprocessing import StandardScaler
sc1=StandardScaler()
x_olcekli = sc1.fit_transform(X) # X verisini scale et

sc2=StandardScaler()
y_olcekli = np.ravel(sc2.fit_transform(Y.reshape(-1,1))) # Y verisini scale et

# SVR Modeli Kur
from sklearn.svm import SVR
svr_reg = SVR(kernel='rbf') # RBF kernel = Esnek eğri
svr_reg.fit(x_olcekli,y_olcekli) # Modele fit et

plt.scatter(x_olcekli,y_olcekli,color='red')
plt.plot(x_olcekli,svr_reg.predict(x_olcekli),color='blue')
plt.title("Support Vector Regression (RBF): Tecrübeye Göre Maaş Tahmini")
plt.xlabel("Ölçeklenmiş Tecrübe")
plt.ylabel("Ölçeklenmiş Maaş")
plt.show()
# 9. SVR ile Tahmin
#print(svr_reg.predict([[11]]))
#print(svr_reg.predict([[6.6]]))
#SVR modeli ölçeklenmiş modellerle eğitildiği için x değerini önce ölçekleyip sonra predict yapmak daha mantıklı!

# X değerini ölçekle
x_test = sc1.transform([[11]]) # Tahmin yapılacak x değeri scale ediliyor
svr_pred_11 = sc2.inverse_transform(svr_reg.predict(x_test).reshape(-1,1))[0][0] # Y tahmini ters scale

x_test2 = sc1.transform([[6.6]])
svr_pred_66 = sc2.inverse_transform(svr_reg.predict(x_test2).reshape(-1,1))[0][0]

print("SVR (RBF kernel) ile 11 yıl tecrübeye sahip birinin tahmini maaşı: {:.2f} TL".format(svr_pred_11))
print("SVR (RBF kernel) ile 6.6 yıl tecrübeye sahip birinin tahmini maaşı: {:.2f} TL".format(svr_pred_66))
```

# Hangi Tahmin Daha Doğru?
### Model kurmak kolaydır, ama doğru model seçimi işin mühendislik kısmıdır.
- Gerçek maaşlar (Y) ile tahmin edilen maaşlar (ŷ) arasındaki farkları ölçmemiz gerekir.
- Buna hata (error) diyoruz.

## En Yaygın Ölçütler (Regresyon Hata Metrikleri):
1. **MAE (Mean Absolute Error) - Ortalama Mutlak Hata**
    - Gerçek ile tahmin arasındaki farkların mutlak değerlerinin ortalamasıdır. Daha düşük = daha iyi.
2. **MSE (Mean Squared Error) – Ortalama Kare Hata**
    - Hataların karesi alınır. Büyük hatalar daha fazla cezalandırılır. Aşırı sapmaları önemser.
3. **RMSE (Root Mean Squared Error) – Karekök Ortalama Hata**
    - MSE'nin karekökü. Yorumlaması daha kolay çünkü birim aynı (TL vs).
4. **R² Score (R-kare) – Determinasyon Katsayısı**
    - 1'e ne kadar yakınsa o kadar iyi.
    - Modelin veriyi ne kadar açıkladığını gösterir.


```python
# Not: Bu çalışmada model davranışlarını ve grafik çizimlerini gözlemlemek amacıyla tüm veriler eğitim (training) için kullanılmıştır.
# Hesaplayacağımız metriklerin amacı: "Model kendi ezberlediği veri üzerinde ne kadar başarılı?"
# Yani kendine karşı sınava girdi gibi düşünebiliriz

# Kural
# - R² yüksek olmalı (ideal 0.8-0.9 üstü)
# - MAE, MSE, RMSE düşük olmalı
# - Train ve Test sonuçları birbirine yakın olmalı

# HATA METRİKLERİ HESAPLAMA
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score

# LINEAR REGRESSION
y_pred_lin = lin_reg.predict(X)
print("Linear Regression:")
print("MAE:", mean_absolute_error(Y, y_pred_lin))
print("MSE:", mean_squared_error(Y, y_pred_lin))
print("RMSE:", np.sqrt(mean_squared_error(Y, y_pred_lin)))
print("R^2:", r2_score(Y, y_pred_lin))
print()

# POLYNOMIAL REGRESSION (derece=4)
y_pred_poly = lin_reg2.predict(poly_reg.fit_transform(X))
print("Polynomial Regression (Derece=4):")
print("MAE:", mean_absolute_error(Y, y_pred_poly))
print("MSE:", mean_squared_error(Y, y_pred_poly))
print("RMSE:", np.sqrt(mean_squared_error(Y, y_pred_poly)))
print("R^2:", r2_score(Y, y_pred_poly))
print()

# SVR
y_pred_svr = svr_reg.predict(x_olcekli)
y_pred_svr_inverse = sc2.inverse_transform(y_pred_svr.reshape(-1,1))
print("SVR (RBF):")
print("MAE:", mean_absolute_error(Y, y_pred_svr_inverse))
print("MSE:", mean_squared_error(Y, y_pred_svr_inverse))
print("RMSE:", np.sqrt(mean_squared_error(Y, y_pred_svr_inverse)))
print("R^2:", r2_score(Y, y_pred_svr_inverse))

# TODO: Train-Test Split yapılacak. 
# Neden train_test_split ? Çünkü ben bu modeli geliştirdim, acaba yeni veri gelse bu model tahmin yapabilir mi test etmem gerekiyor.
# TODO: Cross-Validation yapılacak.
```
