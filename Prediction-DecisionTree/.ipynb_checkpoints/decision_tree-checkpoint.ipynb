# Karar Ağacı ile Tahmin (Decision Tree)
- DecisionTreeRegressor, verileri bölerek (dallanarak) öğrenir.
- 

```python
# MODEL KURULUM - GRAFİK - TAHMİN
# 1. Kütüphaneler
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd

# 2. Veri Yükleme
veriler = pd.read_csv('maaslar.csv')

# X = Bağımsız Değişken (Tecrübe), Y = Bağımlı Değişken (Maaş)
# X ve Y NumPy Array
# x ve y DataFrame
x = veriler.iloc[:,1:2] # Tecrübe sütunu
y = veriler.iloc[:,2:] # Maaş sütunu
X = x.values
Y = y.values

print(x)
print(y)

# 3. Decision Tree
from sklearn.tree import DecisionTreeRegressor
r_dt = DecisionTreeRegressor(random_state=0)
r_dt.fit(X,Y)

# Z ve K, verinin biraz sağından ve solundan alınarak "karar sınırlarını" göstermek için kullanılır. 
# Bu çizgiler sayesinde Decision Tree'nin "basamaklı" yapısı net görünür.
Z = X + 0.5
K = X - 0.5
plt.scatter(X,Y,color='red')
plt.plot(X,r_dt.predict(X),color='blue')

plt.plot(x,r_dt.predict(Z), color='green')
plt.plot(x,r_dt.predict(K), color='yellow')
print(r_dt.predict([[11]]))
print(r_dt.predict([[6.6]]))
```

## Boy ve Kilo Bilgisi Kullanarak Yaş Tahmini


```python
# MODEL KURULUM - GRAFİK - TAHMİN
# 1. Kütüphaneler
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd

# 2. Veri Yükleme
veriler = pd.read_csv('veriler.csv')

# X = Bağımsız Değişken (Tecrübe), Y = Bağımlı Değişken (Maaş)
# X ve Y NumPy Array
# x ve y DataFrame
x = veriler.iloc[:,1:3] # Boy ve Kilo sütunu
y = veriler.iloc[:,3:4] # Yaş sütunu

X = x.values
Y = y.values

print(x)
print(y)

# 3. Decision Tree
from sklearn.tree import DecisionTreeRegressor
r_dt = DecisionTreeRegressor(random_state=0)
r_dt.fit(X,Y)

print(r_dt.predict([[170, 65]])) # Örnek Tahmin

# Grafik (Boy sabit, kilo değişken)
kilo_araligi = np.arange(50, 90, 1).reshape(-1, 1)
tahminler = r_dt.predict(np.hstack((np.full_like(kilo_araligi, 170), kilo_araligi)))

plt.plot(kilo_araligi, tahminler)
plt.xlabel("Kilo")
plt.ylabel("Yaş Tahmini")
plt.title("Boy 170 iken kiloya göre yaş tahmini")
plt.show()
```
