```python
import math

def plot_all_histograms(df, title_prefix=""):
    num_cols = df.select_dtypes(include=[np.number]).columns
    n_cols = 3
    n_rows = math.ceil(len(num_cols) / n_cols)

    plt.figure(figsize=(5 * n_cols, 4 * n_rows))

    for i, col in enumerate(num_cols, 1):
        plt.subplot(n_rows, n_cols, i)
        sns.histplot(df[col], kde=True, bins=30)
        plt.title(f"{title_prefix}{col}")
        plt.xlabel("")
        plt.ylabel("")

    plt.tight_layout()
    plt.show()
```

```python
plot_all_histograms(df)
```

```python
from sklearn.metrics import calinski_harabasz_score, davies_bouldin_score
```

```python
df = pd.read_csv("mall_customers.csv")
df = df.drop("CustomerID",axis=1)
le = LabelEncoder()
df['Gender'] = le.fit_transform(df['Gender'])

features_2d = ["Annual Income (k$)","Spending Score (1-100)"]
features_3d = ["Age", "Annual Income (k$)","Spending Score (1-100)"]
features_4d = ["Gender", "Age", "Annual Income (k$)","Spending Score (1-100)"]

for feats in [features_2d, features_3d, features_4d]:
    X = df[feats]
    X_scaled = MinMaxScaler().fit_transform(X)

    hc = AgglomerativeClustering(n_clusters=5)
    y_hc = hc.fit_predict(X_scaled)

    # Silhouette skoru, küme sayısıı seçiminde (örneğin K-Means) yol gösterici olabilir.
    sil = silhouette_score(X_scaled, y_hc) # 1'e ne kadar yakınsa o kadar iyi, negatif skorlar yanlış kümelenmeye işaret eder.
    # Daha düşük Davies Bouldin Skoru daha iyi kümelenme demektir
    db = davies_bouldin_score(X_scaled, y_hc) # Kümeler arası benzerliği değerlendirerek bir kümeleme modelinin ne kadar ayrık olduğunu ölçer.
    ch = calinski_harabasz_score(X_scaled, y_hc) # Yüksek olması

    print(f"\n features: {feats}")
    print("Silhoutte score: ",sil) # -1 ila 1, yüksek olması tercih edilir -> Ayrım + sıkılık
    print("Davies Bouldin score: ", db) # [0,∞), düşük olması tercih edilir -> Küme benzerliği
    print("Calinski Harabasz score: ", ch) # [0,∞), yüksek olması tercih edilir. -> Varyans oranı
    print("---------------")
```

```python
from sklearn.cluster import KMeans
df = pd.read_csv("mall_customers.csv")
df = df.drop("CustomerID", axis=1)
le = LabelEncoder()
df['Gender'] = le.fit_transform(df['Gender'])

features_2d = ["Annual Income (k$)","Spending Score (1-100)"]
features_3d = ["Age", "Annual Income (k$)","Spending Score (1-100)"]
features_4d = ["Gender", "Age", "Annual Income (k$)","Spending Score (1-100)"]

for feats in [features_2d, features_3d, features_4d]:
    X = df[feats] 
    X_scaled = MinMaxScaler().fit_transform(X)

    kmeans = KMeans(n_clusters=5)
    y_hc = kmeans.fit_predict(X_scaled)

    sil = silhouette_score(X_scaled,y_hc)
    db = davies_bouldin_score(X_scaled,y_hc)
    ch = calinski_harabasz_score(X_scaled,y_hc)

    print(f"\n features : {feats}")
    print(f"Silhoutte score: {sil}")
    print(f"Davies Bouldin score: {db}")
    print(f"Calinski Harabasz score: {ch}")
    print("-------------------------")
```

## MinMaxScaler & StandardScaler

|Özellik |MinMaxScaler |StandardScaler |
|:-----|:----:|----:|
|Amaç |Veriyi belirli bir aralığa sıkıştırmak| Veriyi normalize edip ortalamasını 0, standart sapmasını 1 yapmak|
|Formül |(x-min) / (max-min) -> [0,1] aralığı (default) |(x - mean) / std |
|Duyarlılık |Aykırı değerlere (outlier) çok duyarlı |Aykırı değerlere karşı biraz daha dayanıklı|

__Örnek:__ Diyelim ki bir alışveriş sitesinde kullanıcıların boyu ve kilosu alınıyor.
- Boylar 150 cm - 200 cm arası, Kilolar 40 kg - 130 kg arası
### MinMaxScaler
Bu yöntemde:
- En kısa boy: 150 -> 0 olur
- En uzun boy: 200 -> 1 olur
- 175 cm biri: (175-150)/(200-150) = 0.5

Yani her değeri 0-1 arasına sıkıştırır.
Ama eğer biri 300 cm çıkarsa, tüm denge bozulur çünkü 1'in üzerine çıkar.
### StandardScaler
- Tüm boy değerlerinin ortalaması alınır.
- Standard sapma ile normalize edilir.
- Sonuç: Ortalama=0, Standart Sapma=1
- Bu da "z-puanı" mantığıdır. Değerin ne kadar ortalamadan sapma gösterdiğini verir.
