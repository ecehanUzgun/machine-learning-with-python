## **Model Eğitimi ve Değerlendirme**
   - Aşağıdaki sınıflandırma algoritmaları uygulanmıştır:
     - Logistic Regression  
     - K-Nearest Neighbors (KNN)  
     - Support Vector Machine (SVM)  
     - Naive Bayes  
     - Decision Tree  
     - Random Forest  
   - Modeller `confusion matrix`, `accuracy`, `precision`, `recall`, `f1-score` gibi metriklerle değerlendirilmiştir.

- **Görselleştirme**
   - Her model için `confusion matrix` grafiği `seaborn` kütüphanesi ile çizilmiştir.

```python
# ImportError: Missing optional dependency 'xlrd'. Install xlrd >= 2.0.1 for xls Excel support Use pip or conda to install xlrd.
!pip install xlrd
# ! işareti Jupyter içinde terminal komutlarını çalıştırmak için kullanılır.
# xlrd, .xls dosyasını okumayı sağlayan yardımcı Python paketidir.
```

```python
# Veri Hazırlığı Fonksiyonu
def load_and_prepare_data(path='Iris.xls'):
    import pandas as pd
    from sklearn.model_selection import train_test_split
    from sklearn.preprocessing import StandardScaler
    
    data = pd.read_excel(path)
    
    X = data.iloc[:, 1:4].values
    y = data.iloc[:,4].values # values -> 1D vektör

    X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.33, random_state=0)
    sc = StandardScaler()
    X_train = sc.fit_transform(X_train)
    X_test = sc.transform(X_test) 

    return X_train, X_test, y_train, y_test
```

```python
# Model Eğitim ve Değerlendirme Fonksiyonu
def train_and_evaluate_model(model, X_train, X_test, y_train, y_test, model_name="Model"):
    from sklearn.metrics import confusion_matrix, accuracy_score, classification_report
    import seaborn as sns #seaborn, Python'da veri görselleştirme için kullanılan güçlü ve yüksek seviyeli bir kütüphanedir. Matplotlib üzerine inşa edilmiştir.
    # seaborn, özellikle istatistiksel grafikler için sade ve estetik görseller üretir.
    import matplotlib.pyplot as plt
    import numpy as np

    model.fit(X_train, y_train)
    y_pred = model.predict(X_test)

    print(f"\n {model_name}")
    print("Accuracy:", accuracy_score(y_test, y_pred))
    print("Classification Report:\n", classification_report(y_test, y_pred))
    # Sınıf isimleri
    #class_names = ['Iris-setosa','Iris-versicolor','Iris-virginica']
    class_names = np.unique(y_test)
    # Confusion Matrix
    cm = confusion_matrix(y_test, y_pred, labels=class_names)
    sns.heatmap(cm, annot=True, fmt = "d", cmap="Blues", xticklabels=class_names, yticklabels=class_names)
    plt.title(f"{model_name} - Confusion Matrix")
    plt.xlabel("Predicted - Tahmin Edilen Etiket")
    plt.ylabel("Actual - Gerçek Etiket")
    plt.show()
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.neighbors import KNeighborsClassifier
from sklearn.svm import SVC
from sklearn.naive_bayes import GaussianNB
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier

X_train, X_test, y_train, y_test = load_and_prepare_data()
```

## Accuracy 
***Accuracy = Doğru Tahmin Sayısı / Toplam Tahmin Sayısı***
Yani, Modelin doğru sınıflandırdığı örneklerin sayısı, toplam tahmin edilen örnek sayısına bölünür.
- __Avantajı:__ Hızlı ve genel başarıyı ölçmek için kullanılır.
- __Dezavantajı:__ Dengesiz veri setlerinde yanıltıcı olabilir.
- __Ne zaman kullanılır?__
    - Sınıflar dengeliyse (örnek sayıları birbirine yakınsa) accuracy genellikle güvenlidir.
    - Daha adil değerlendirme için Precision, Recall, F1-score gibi metriklerle birlikte değerlendirilmelidir.

## Classification Report
- Her sınıf için modelin ne kadar iyi çalıştığını ölçen 4 temel metrik verir:
| **Metrik**     | **Açıklama**                                         | **Ne Anlama Geliyor**                                            |
|----------------|------------------------------------------------------|-------------------------------------------------------------------|
| **Precision**  | Doğru pozitif / (Doğru pozitif + Yanlış pozitif)    | Model `Iris-setosa` dediğinde ne kadar isabetliydi?              |
| **Recall**     | Doğru pozitif / (Doğru pozitif + Yanlış negatif)    | Gerçekten `Iris-setosa` olanları model ne kadar bulabildi?       |
| **F1-Score**   | Precision ve Recall’un harmonik ortalaması           | Dengeyi gösterir. İkisinin ortalamasıdır.                         |
| **Support**    | O sınıfa ait test örneği sayısı                      | Gerçekten test setinde kaç tane örnek vardı.                      |


## Logistic Regression

```python
train_and_evaluate_model(LogisticRegression(), X_train, X_test, y_train, y_test, "Logistic Regression")
```

## Confusion Matrix 
- Bu matris, modelin hangi sınıflarda doğru/yanlış tahmin yaptığını görselleştirir.
    - ***Satırlar*** Gerçek Sınıf
    - ***Sütunlar*** Tahmin Edilen Sınıf
    - ***Diagonal (köşegen)*** Doğru Tahminler
- Iris-setosa (0): 16 örnek -> Hepsi doğru sınıflanmış
- Iris-versicolor (1): 19 örnek -> 18'i doğru, 1'i yanlışlıkla virginica olarak sınıflanmış
- Iris-virginica (2): 15 örnek -> 13'ü doğru, 2'si versicolor olarak tahmin edilmiş

## KNN

```python
train_and_evaluate_model(KNeighborsClassifier(n_neighbors=3), X_train, X_test, y_train, y_test, "KNN (k=3)")
```

## SVM

```python
train_and_evaluate_model(SVC(kernel='poly'), X_train, X_test, y_train, y_test, "SVM (Poly Kernel)")
```

## Naive Bayes

```python
train_and_evaluate_model(GaussianNB(), X_train, X_test, y_train, y_test, "Naive Bayes")
```

## Decision Tree

```python
train_and_evaluate_model(DecisionTreeClassifier(criterion='entropy'), X_train, X_test, y_train, y_test, "Decision Tree")
```

## Random Forest

```python
train_and_evaluate_model(RandomForestClassifier(n_estimators=10), X_train, X_test, y_train, y_test, "Random Forest")
```
