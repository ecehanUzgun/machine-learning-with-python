```python
# libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.metrics import confusion_matrix
```

```python
veriler = pd.read_excel('Iris.xls')
print(veriler.head())
```

```python
x = veriler.iloc[:,1:4].values #baÄŸÄ±msÄ±z deÄŸiÅŸkenler
```

```python
y = veriler.iloc[:,4:].values #baÄŸÄ±mlÄ± deÄŸiÅŸken
```

```python
y_flat = y.ravel()
counts = pd.Series(y_flat).value_counts()
print(counts)
```

```python
counts.plot(kind='pie', autopct='%1.1f%%', startangle=90)
plt.title("Iris Data Set")
plt.ylabel("")
plt.show
```

## Train/Test Split
- Verilerin EÄŸitim ve Test iÃ§in bÃ¶lÃ¼nmesi

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x,y,test_size=0.33, random_state=0)
```

## Feature Scaling
- Verilerin Ã–lÃ§eklenmesi

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
X_train = sc.fit_transform(x_train)
X_test = sc.fit_transform(x_test)
```

# SÄ±nÄ±flandÄ±rma AlgoritmalarÄ±

1. Logistic Regression

```python
from sklearn.linear_model import LogisticRegression
logr = LogisticRegression(random_state=0)
logr.fit(X_train,y_train.ravel()) #egitim - ravel() Ã§ok boyutlu diziyi tek boyuta indirger. 
y_pred = logr.predict(X_test) #tahmin
#Confusion Matrix
from sklearn.metrics import confusion_matrix
cmLogisticReg = confusion_matrix(y_test,y_pred)
print(cmLogisticReg)
```

2. KNN

```python
from sklearn.neighbors import KNeighborsClassifier
knn = KNeighborsClassifier(n_neighbors=1, metric='minkowski')
knn.fit(X_train,y_train.ravel())
# DataConversionWarning: A column-vector y was passed when a 1d array was expected. Please change the shape of y to (n_samples,), for example using ravel().
y_pred_knn = knn.predict(X_test)
cm_knn = confusion_matrix(y_test,y_pred_knn)
print(cm_knn)
```

3. SVC

```python
from sklearn.svm import SVC
svc = SVC(kernel='poly') # TODO:rbf, linear, sigmoid gibi kernel'lar da denenecek.
svc.fit(X_train,y_train.ravel())
y_pred_svc = svc.predict(X_test)
cm_svc = confusion_matrix(y_test,y_pred_svc)
print(cm_svc)
```

4. Naive Bayes

```python
from sklearn.naive_bayes import GaussianNB
gnb = GaussianNB()
gnb.fit(X_train,y_train.ravel())
y_pred_nb = gnb.predict(X_test)
cm_nb = confusion_matrix(y_test,y_pred_nb)
print(cm_nb)
```

5. Decision Tree

```python
from sklearn.tree import DecisionTreeClassifier
dtc = DecisionTreeClassifier(criterion='entropy')
dtc.fit(X_train,y_train)
y_pred_dt=dtc.predict(X_test)
cm_dt = confusion_matrix(y_test,y_pred_dt)
print(cm_dt)
```

6. Random Forest

```python
from sklearn.ensemble import RandomForestClassifier
rfc = RandomForestClassifier(n_estimators=10, criterion='entropy')
rfc.fit(X_train,y_train.ravel())
y_pred_rfc = rfc.predict(X_test)
cm_rf = confusion_matrix(y_test,y_pred_rfc)
print(cm_rf)
```
